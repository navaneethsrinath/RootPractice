```python
import ROOT
inFileName = "data/fardet_data_2_run44299_R20-11-25-cosfilt.c_r44299.0-r44299.53.cosmic_ntuple.root"
inFile = ROOT.TFile.Open(inFileName, "READ")
tree = inFile.Get("cosmicntuple/cosmicTree")
canv = ROOT.TCanvas("Canvas", "NewGraph", 1920, 1080)
```

```python
# X-Position Definitions
xstartPos = ROOT.TH1D("startposx","Starting X Position", 375, -1000, 1000)
xstopPos = ROOT.TH1D("stopposx", "Stopping X Position", 375, -1000, 1000)
# Y-Position Definitions
ystartPos = ROOT.TH1D("startposY", "Starting Y Position", 250, -1000, 1000)
ystopPos = ROOT.TH1D("stopposY", "Stopping Y Position", 375, -1000, 1000)
# Z-Position Definitions
zstartPos = ROOT.TH1D("startposZ","Starting Z Position", 1000, 0, 6000)
zstopPos = ROOT.TH1D("stopposZ", "Stopping Z Position", 1000, 0, 6000)
# X-Direction Definitions
xStartDir = ROOT.TH1D("startdirX", "Starting X Direction", 250, -1, 1)
xStopDir = ROOT.TH1D("stopdirX", "Stopping X Direction", 250, -1, 1)
# Y-Direction Definitions
yStartDir = ROOT.TH1D("startdirY", "Starting Y Direction", 250, -1, 1)
yStopDir = ROOT.TH1D("stopdirY", "Stopping Y Direction", 250, -1, 1)
# Z-Direction Definitions
zStartDir = ROOT.TH1D("startdirZ", "Starting Z Direction", 250, -1, 1)
zStopDir = ROOT.TH1D("stopdirZ", "Stopping Z Direction", 250, -1, 1)
# Energy
totalGeV = ROOT.TH1D("totalGeV", "Total Energy", 250, 0, 4)
# Length
totalLength = ROOT.TH1D("totalLength", "Total Length", 250, 0, 3000)
```

```python
for entryNum in range(0, tree.GetEntries()):
    tree.GetEntry(entryNum)
    if getattr(tree, "ntrack") == 0:
        continue
    # X-position stuff
    startX = getattr(tree, "startPosX")[0]
    stopX = getattr(tree, "stopPosX")[0]
    xstartPos.Fill(startX)
    xstopPos.Fill(stopX)
    # Y-position stuff
    startY = getattr(tree, "startPosY")[0]
    stopY = getattr(tree, "stopPosY")[0]
    ystartPos.Fill(startY)
    ystopPos.Fill(stopY)
    # Z-position stuff
    startZ = getattr(tree, "startPosZ")[0]
    stopZ = getattr(tree, "stopPosZ")[0]
    zstartPos.Fill(startZ)
    zstopPos.Fill(stopZ)
    # X-Direction stuff
    startdirectionX = getattr(tree, "startDirX")[0]
    stopdirectionX = getattr(tree, "stopDirX")[0]
    xStartDir.Fill(startdirectionX)
    xStopDir.Fill(stopdirectionX)
    # Y-Direction stuff
    startdirectionY = getattr(tree, "startDirY")[0]
    yStartDir.Fill(startdirectionY)
    stopdirectionY = getattr(tree, "stopDirY")[0]
    yStopDir.Fill(stopdirectionY)
    # Z-Direction stuff
    startdirectionZ = getattr(tree, "startDirZ")[0]
    stopdirectionZ = getattr(tree, "stopDirZ")[0]
    zStartDir.Fill(startdirectionZ)
    zStopDir.Fill(stopdirectionZ)
    # Energy
    GeV = getattr(tree, "totalGeV")[0]
    totalGeV.Fill(GeV)
    # Length
    Length = getattr(tree, "totalLength")[0]
    totalLength.Fill(Length)

    
```

```python
xstartPos.Draw()
xstartPos.GetYaxis().SetTitle("Occurances")
xstartPos.GetXaxis().SetTitle("Distance From Center (cm)")
xstartPos.SetLineColor(ROOT.kRed)
canv.Draw()
```

```python
LengthGeV = ROOT.TH2D("lengthGeV", "Length vs Energy", 350, 0, 2.5, 350, 0, 3000)


for entryNum in range(0, tree.GetEntries()):
    tree.GetEntry(entryNum)
    if getattr(tree, "ntrack") == 0:
        continue
    if getattr(tree, "startPosY")[0] < 750:
        continue
    if getattr(tree, "totalGeV")[0] > 2.5:
        continue
    if getattr(tree, "startPosY")[0] < (-750):
        continue
    Length = getattr(tree, "totalLength")[0]
    GeV = getattr(tree, "totalGeV")[0]
    LengthGeV.Fill(GeV, Length)

LengthGeV.GetYaxis().SetTitle("Length (cm)")
LengthGeV.GetXaxis().SetTitle("Energy (GeV)")
ROOT.gStyle.SetPalette(105)

LengthGeV.Draw("colz")
canv.Update()
canv.Draw()


# Define the fit function
fitFunction = ROOT.TF1("fitFunction", "[0] + [1]*x", 0, 2.5)  # y = c + mx, assuming linear fit

# Perform the fit
fitResult = LengthGeV.Fit(fitFunction, "S")  # "S" option to silence fit output

# Access fit parameters
slope = fitResult.Parameter(1)
intercept = fitResult.Parameter(0)

# Draw the best-fit line
fitFunction.Draw("same")

# Update the canvas and draw it
canv.Update()
canv.Draw()
```
